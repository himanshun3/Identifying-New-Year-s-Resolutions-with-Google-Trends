```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
% matplotlib inline
sns.set()
```

```python
#Using skiprows to skip the first row at the start of the file

df=pd.read_csv('multiTimeline.csv',skiprows=1)

#Renaming Columns so that they do not have whitespaces in between them

df.columns=['Month','Diet','Gym','Finance']
df.head()
```

```python
df.info()
```

```python
#Changing MOnth column into a datetime data type and setting it index

df['Month']=pd.to_datetime(df['Month'])
df.set_index('Month',inplace=True)
```

```python
df.head()
```

```python
#Exploratory Data Analysis

df.plot(figsize=(20,10),fontsize=20,linewidth=5)
plt.xlabel('Year',fontsize=20);
```

```python
#Plotting only the diet column

df[['Diet']].plot(figsize=(20,10),fontsize=20,linewidth=5)
plt.xlabel('Year',fontsize=20);
```

```python
#Taking the rolling average of the Diet COlumn
Diet=df[['Diet']]

#Taking Window size of 12 months

Diet.rolling(12).mean().plot(figsize=(20,10),linewidth=5,fontsize=20)
plt.xlabel('Year',fontsize=20);
```

```python
#Taking Rolling Average of the Gym Column

Gym=df[['Gym']]

#Taking Window size of 12 months

Gym.rolling(12).mean().plot(figsize=(20,10),linewidth=5,fontsize=20,color='Red')
plt.xlabel('Year',fontsize=20);
```

```python
#Comparing both the gym and Diet Rolling average graphs

df_rm=pd.concat([Diet.rolling(12).mean(),Gym.rolling(12).mean()],axis=1)
df_rm.plot(linewidth=5,figsize=(20,10),fontsize=20)
plt.xlabel('Year',fontsize=20);
```

```python
Diet.diff().plot(linewidth=5,fontsize=20,figsize=(20,10))
plt.xlabel('Year',fontsize=20);
```

```python
df.plot(figsize=(20,10),linewidth=5,fontsize=20)
plt.xlabel('Year',fontsize=20);
```

```python
#Plotting the first order difference 

df.diff().plot(figsize=(20,10),fontsize=20,linewidth=5)
plt.xlabel('Year',fontsize=20);
```
